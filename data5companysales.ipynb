```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df=pd.read_csv("company_sales_data.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
#read total profit of all months an show it in a line graph
#line style=doted, clr=green ,show legend at lower right,add a title,add a circle marker,line width should be 3
x=df["month_number"]
y=df["total_profit"]
plt.plot(x,y,color="green",linestyle="dotted",marker='X',linewidth=3,markerfacecolor="red",ms=10)       #ms=marker size
plt.xlabel("month_number")
plt.ylabel("total_profit")
plt.title("month profit graph")
plt.xticks(x,["jan","feb","mar","apr","may","june","jul","aug","sep","oct","nov","dec"])
plt.legend("company profit",loc="lower right")
plt.show()
```

```python
#read all product sales data and show it in a multiline plot

x=df["month_number"]
y1=df["facecream"]
y2=df["facewash"]
y3=df["toothpaste"]
y4=df["bathingsoap"]
y5=df["shampoo"]
y6=df["moisturizer"]
df.plot(x="month_number",y=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"],linewidth=4,marker='o',markerfacecolor="black")
plt.show()
```

```python
x=df["month_number"]
y=df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']]
df.plot(x="month_number",y=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"])
plt.show()
```

```python
#read toothpaste sales data of each month and show it using scatter plot with grid lines
x=df["month_number"]
y=df["toothpaste"]
plt.scatter(x,y,color="red")
plt.grid(True)
plt.xticks(x,["jan","feb","mar","apr","may","june","jul","aug","sep","oct","nov","dec"])
plt.show()
```

```python
#read facecream and facewash product and show it ina bargraph


x=df["month_number"]
y1=df["facecream"]
y2=df["facewash"]
plt.bar(x-.2,y1,width=0.4,color="purple")
plt.bar(x+.2,y2,width=0.4,color="violet")
plt.xlabel("month")
plt.ylabel("products")
plt.show()

```

```python
x=df["month_number"]
y=df[["facecream","facewash"]]
df.plot(x="month_number",y=["facecream","facewash"],kind="bar",color=['violet','skyblue'],grid=True)

plt.show()
```

```python
#read the total profit of each month and show it using a histogram and most common profit range

x=df["total_profit"]
y=[100000,200000,300000,400000,500000]
plt.hist(x,y)
plt.show()
```

```python
#calculate total sales data in last year for each product using pie chart
y=[df["facecream"].sum(),df["facewash"].sum(),df["toothpaste"].sum(),df["bathingsoap"].sum(),df["shampoo"].sum(),df["moisturizer"].sum()]
y
labels=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
plt.pie(y,labels=labels,autopct='%1.1f%%')
plt.legend()
plt.show()
```
