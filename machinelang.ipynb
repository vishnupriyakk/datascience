```python
x=[1,2,4,6,7,10,12]
y=[120,115,114,113,111,90,95]                          #m=slop,c=interseption
m,c,r,p,std_err=stats.linregress(x,y)
def fun(x):
    return(m*x)+c


p=list(map(fun,x))
plt.scatter(x,y)
plt.plot(x,p)
plt.show()
```
