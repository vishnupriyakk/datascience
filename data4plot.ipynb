```python
x=np.array([3,4,5,6,2])
y1=[6,7,3,4,1]
y2=[8,5,6,3,4]
plt.bar(x-.2,y1,width=0.4)
plt.bar(x+.2,y2,width=0.4)
plt.show()
```

```python
#groupchart
x=np.array([3,4,5,6,2])
y1=[6,7,3,4,1]
y2=[8,5,6,3,4]
y3=[4,6,5,2,3]


plt.bar(x-.2,y1,width=0.4,color="purple")
plt.bar(x+.2,y2,width=0.4,color="violet")
plt.title("GRAPH")
plt.legend(["a","b","c"])
plt.xlabel("duration1")
plt.ylabel("duration2")
plt.xticks(x,["first","second","third","fourth","fifth"])
plt.show()
```

```python
#stack chart
x=["a","b","c","d"]
y1=[3,5,6,7]
y2=[5,6,7,4]
plt.bar(x,y1,width=0.4,color="violet")
plt.bar(x,y2,width=0.4,color="purple",bottom=y1)
plt.show()
```

```python
x=["a","b","c","d"]
y1=np.array([3,5,4,6])
y2=np.array([5,6,7,4])
y3=np.array([8,5,6,7])
y4=np.array([5,7,9,3])
plt.bar(x,y1,width=0.4,color="violet")
plt.bar(x,y2,width=0.4,color="purple",bottom=y1)
plt.bar(x,y3,width=0.4,color="skyblue",bottom=y1+y2)
plt.bar(x,y4,width=0.4,color="blue",bottom=y1+y2+y3)

plt.show()
```
